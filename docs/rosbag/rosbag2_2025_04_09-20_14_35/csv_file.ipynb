# Odometry Log Analysis


## Observations

### 1. Trajectory (X vs Y)
- The robot followed a relatively **straight diagonal path** from approximately (x  9, y  0) to (x  -40, y  27).
- No significant loops or erratic movements were observed during the initial movement phase.
- Indicates **forward motion with minor steering**, possibly along a corridor.

### 2. Position Over Time
- Both X and Y positions remained **constant for the first 400500 seconds**, indicating the robot was either idle or waiting.
- A **sudden jump** in position occurred after ~400s, suggesting:
  - A relocalization event (e.g., AMCL or SLAM correction).
  - A significant correction or reset in the localization system.

### 3. Linear Velocity (x)
- Unrealistically high values observed (5001900 m/s), far beyond the expected range for a mobile robot.
- Velocity remained stable only after fixing the velocity calculation logic.
- Indicates the odometry node was incorrectly using **total cumulative distance** instead of **step-wise delta distance**.

### 4. Yaw Angle (Orientation over Time)
- The yaw increased steadily, suggesting slow **counter-clockwise rotation** during movement.
- Sharp discontinuities and wraparound were observed ( to -), typical of angular wrapping.
- After ~1100s, yaw became **noisy and unstable**, possibly due to:
  - Localization correction.
  - Sensor noise or TF misalignment.

---

## Issues Identified

1. **Velocity Calculation Bug**
   - Velocity was incorrectly calculated as:
     ```
     linear.x = total_distance / dt
     ```
     instead of:
     ```
     linear.x = delta_distance / dt
     ```
   - This resulted in massive, incorrect velocity spikes.

2. **Localization/SLAM Reset**
   - Sudden jump in pose and instability in yaw between 4001200s.
   - Likely caused by a map reset or relocalization event.

3. **Potential TF Drift**
   - If yaw instability is persistent across bags, the TF tree or IMU fusion may need tuning.
   - Consider checking TF tree alignment and IMU frame calibration.

---

## Fixes Applied

- Velocity calculation corrected to use `delta_distance / dt`.
- Code was updated to track step-wise distance properly.
- Future tests will confirm whether the correction stabilizes velocity plots.

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.transform import Rotation as R # For quaternion to Euler conversion

if 'time_elapsed_s' not in df.columns:
    if 'timestamp' in df.columns:
        df = df.sort_values('timestamp').reset_index(drop=True)
        start_time = df['timestamp'].iloc[0]
        df['time_elapsed_s'] = (df['timestamp'] - start_time) * 1e-9 # Convert ns to s
    else:
        print("Error: 'timestamp' column not found, cannot calculate time.")
       

# --- 1. Extract Quaternions ---
# Scipy expects quaternions in [x, y, z, w] order
quaternions = df[['orient_x', 'orient_y', 'orient_z', 'orient_w']].values

# --- 2. Convert Quaternions to Euler Angles (including Yaw) ---
# Create Rotation objects from the quaternions
rotations = R.from_quat(quaternions)

# Convert to Euler angles. 'zyx' means the rotations are applied in
# Z (Yaw), then Y (Pitch), then X (Roll) order.
# The output is [yaw, pitch, roll] in radians.
euler_angles_rad = rotations.as_euler('zyx', degrees=False)

# --- 3. Extract Yaw and Add to DataFrame ---
# Yaw is the first angle in the 'zyx' sequence
df['yaw_rad'] = euler_angles_rad[:, 0]
df['yaw_deg'] = np.degrees(df['yaw_rad']) # Also store in degrees if needed

# --- 4. Plot Yaw vs. Time ---

# Plot Yaw in Radians
plt.figure(figsize=(12, 6))
plt.plot(df['time_elapsed_s'], df['yaw_rad'])
plt.xlabel("Time Elapsed (s)")
plt.ylabel("Yaw Angle (radians)")
plt.title("Yaw Angle over Time")
plt.grid(True)
# Optional: Set y-limits for radians if desired (e.g., -pi to pi)
# plt.ylim(-np.pi, np.pi)
plt.show()

# Optional: Plot Yaw in Degrees
plt.figure(figsize=(12, 6))
plt.plot(df['time_elapsed_s'], df['yaw_deg'])
plt.xlabel("Time Elapsed (s)")
plt.ylabel("Yaw Angle (degrees)")
plt.title("Yaw Angle over Time")
plt.grid(True)
# Optional: Set y-limits for degrees if desired (e.g., -180 to 180)
# plt.ylim(-180, 180)
plt.show()

# You can also look at the angular velocity around Z, which *should*
# correspond to the rate of change of yaw:
plt.figure(figsize=(12, 6))
plt.plot(df['time_elapsed_s'], df['ang_vel_z'], label='Angular Velocity Z (from data)')

# Optional: Calculate yaw rate numerically from calculated yaw angle for comparison
# delta_yaw = df['yaw_rad'].diff()
# delta_time = df['time_elapsed_s'].diff()
# calculated_yaw_rate = delta_yaw / delta_time
# plt.plot(df['time_elapsed_s'], calculated_yaw_rate, label='Yaw Rate (calculated)', linestyle='--')

plt.xlabel("Time Elapsed (s)")
plt.ylabel("Angular Velocity Z (rad/s)")
plt.title("Angular Velocity Z (Yaw Rate) over Time")
plt.legend()
plt.grid(True)
plt.show()
````
